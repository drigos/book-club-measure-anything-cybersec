```python
# Plot a uniform distribution
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Plot the data
def plot_data(data, xlim=None):
    plt.figure(figsize=(16, 4))
    if xlim is not None:
        plt.xlim(xlim)
    plt.hist(data, bins=30)
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.title('Distribution')
    plt.axvline(np.median(data), color='red', linestyle='dashed', linewidth=1)
    plt.axvline(np.mean(data), color='yellow', linestyle='dashed', linewidth=1)
    (_, idx, counts) = np.unique(data, return_index=True, return_counts=True)
    index = idx[np.argmax(counts)]
    mode = data[index]
    plt.axvline(mode, color='purple', linestyle='dashed', linewidth=1)
    plt.show()
    # Clear the plot
    plt.clf()
```

```python
uniformDist = np.random.uniform(0, 1, 10000)
```

```python
plot_data(uniformDist)
```

```python
samples = np.random.choice(uniformDist, 5)
plot_data(samples, (0, 1))
```

```python
# Monte Carlo simulation
# https://en.wikipedia.org/wiki/Monte_Carlo_method
executions = 100000
expected = 0
for i in range(executions):
    samples = np.random.choice(uniformDist, 5)
    if samples.min() < np.median(uniformDist) and samples.max() > np.median(uniformDist):
        expected = expected+1

print(f'{expected} / {executions} ({expected / executions * 100:.2f}%)')
```

```python
normalDist = np.random.normal(0, 1, 10000)
```

```python
plot_data(normalDist, (normalDist.min(), normalDist.max()))
```

```python
samples = np.random.choice(normalDist, 5)
plot_data(samples, (normalDist.min(), normalDist.max()))
```
